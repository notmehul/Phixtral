```python
model_name = "phixtral-2x2_8" # @param ["phixtral-2x2_8", "phixtral-4x2_8"]
instruction = "tell me a story about a frog and a bird" # @param {type:"string"}

prompt = f'''
<|im_start|>system
You are Phixtral, a helpful AI assistant.<|im_end|>
<|im_start|>user
{instruction}<|im_end|>
<|im_start|>assistant
'''
```

```python
import torch

torch.set_default_device("cuda")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

# Load the model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    f"mlabonne/{model_name}",
    torch_dtype="auto",
    load_in_4bit=True,
    trust_remote_code=True
)
tokenizer = AutoTokenizer.from_pretrained(
    f"mlabonne/{model_name}",
    trust_remote_code=True
)
```

```python
# Tokenize the input string
inputs = tokenizer(
    prompt,
    return_tensors="pt",
    return_attention_mask=False
)
```

```python
# Generate text using the model
outputs = model.generate(**inputs, max_length=200)

# Decode and print the output
text = tokenizer.batch_decode(outputs)[0]
print(text[len(prompt):])
```
